# 1D Sattelite Attitude Controller

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
import control as co
```

```python
plt.rcParams['figure.figsize'] = [23,10]
plt.rcParams['font.size'] = 24

s, t = symbols("s t")

J = 100 #1-D Inertia 'Tensor'
Gs = 1/(J*s**2) #The TF in Laplace domain
```

```python
# Consider an impulse reference input (u(t) = delta(t) <=> U(s) = 1)

Us = 1
Ys = Gs*Us

yt = np.laplace(Ys) #Noting that y(t) = g(t) is the impulse response

print(yt)
```

```python
plt.plot(yt)
plt.show()
```
