# For plotting the Probability vs. Number of Heads for EP 400

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
db = pd.read_csv('PvsH.csv')
x = list(db['Num of Heads'])
y = list(db['Probability'])

plt.figure(figsize=(7,5))
plt.style.use('seaborn')
plt.plot(x,y, c='purple')
plt.scatter(x,y,marker="o",s=50,edgecolors="black",c="red")
plt.title("Probablity vs. # of Heads")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.savefig('Prob VS Head')
plt.show()
```
