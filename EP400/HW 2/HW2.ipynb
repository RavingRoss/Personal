```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
from scipy.optimize import fsolve
```

```python
def plot1(n=10):
    x=np.arange(0,30,0.1)
    plt.axhline(color='k')
    plt.axvline(color='k')
    line1,=plt.plot(x,x**n*np.exp(-x))
    line2,=plt.plot(x,n**n*np.exp(-n)*np.exp(-(x-n)**2/(2*n)))
    plt.grid
    plt.xlabel('X',fontsize=12)
    plt.title('n=10')
    plt.legend((line1,line2), ('f(x)','g(x)'))
    return 0
plot1()
plt.savefig('n=10')
```
