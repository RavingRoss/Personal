# Problem 2.11 In CHP 2 for HW6

```python
import matplotlib.pyplot as plt
import pandas as pd
from sympy import *
```

```python
df = pd.read_csv('PB11.csv')

x = list(df[('qA')])
y1 = list(df[('OmegaT')])
y2 = list(df['Prob'])

fig, ax = plt.subplots(figsize=(8,5))
ax2 = ax.twinx()
plt.title("$\Omega_{T}$ vs. q$_{A}$")
ax.scatter(x,y1,marker="o",s=50,edgecolors="black",c="purple")
ax.plot(x, y1, color = 'black')
ax.set_xlabel("Energy of Paramgnet A")
ax.set_ylabel('Total Multiplicity ($\Omega_{T}$)')
ax2.set_ylabel("Probability")
ax.text(x=0,y=1.60e56,s='Peak Probability at q$_{A}$ = q$_{B}$ = 40: 11.471%')
plt.savefig('Prob VS qA')
plt.show()
```

# Problem 2.21 in CHP 2 for HW6

```python
z = symbols('z',real=True)

'''qA = z*q
qB = q(1-z) where l = qA*qB
mult = (q/(2*n)**(2n)*exp(2*n)[4*z(1-z)]**n), ignoring the constant part
'''

f0 = (4*(z-z**2))
f1 = (4*(z-z**2))**10
f2 = (4*(z-z**2))**100
f3 = (4*(z-z**2))**1000
f4 = (4*(z-z**2))**10000

a = plot(f0, (z,0,1) ,ylabel='$[4z(1-z)]^N$',legend=True,line_color='r',title='Plot of the Mulitiplicity function vs z',size=(10,7),show=False)
a.extend(plot(f1,(z,0,1) ,legend=True,line_color='b',show=False))
a.extend(plot(f2,(z,0,1) ,legend=True,line_color='g',show=False))
a.extend(plot(f3,(z,0,1) ,legend=True,line_color='purple',show=False))
a.extend(plot(f4,(z,0,1) ,legend=True,line_color='orange',show=False))
a.save('Mult vs z')
a.show()
```
