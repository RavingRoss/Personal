# This is Lab 06, Random Decay Processes and Shielding Application

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sympy import *
```

# Sodium Data from Lab

```python
# Plot for Sodium
# demonstration.
var = pd.read_csv('JammaRayHistogrammie.csv')
data = np.array(var["Counts"])
# Fit a normal distribution to
# the data:
# mean and standard deviation
mu, std = norm.fit(data)
# Plot the histogram.
plt.hist(data, bins=10, density=True, alpha=0.6, edgecolor='black', color='b')
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
plt.title("Histogram Plot for Radioactive Decay [Sodium]")
plt.xlabel("Counts [400$^{-s}$]")
plt.ylabel("Frequncy of Occurrance")
plt.savefig('Co-Plot')
plt.show()
```

# Cobalt Data from Lab

```python
# Plot for Cobalt
# demonstration.
var = pd.read_csv('JammaRayHistogrammie2.csv')
data = np.array(var["Counts"])
# Fit a normal distribution to
# the data:
# mean and standard deviation
mu, std = norm.fit(data)
# Plot the histogram.
plt.hist(data, bins=15, density=True, alpha=0.6, edgecolor='black', color='r')
# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
plt.title("Histogram Plot for Radioactive Decay [Cobalt]")
plt.xlabel("Counts [400$^{-s}$]")
plt.ylabel("Frequncy of Occurrance")
plt.savefig('Co-Plot')
plt.show()
```

# ln() Plot 

```python
var = pd.read_csv("JammaRayHistogrammie2.csv")
x = np.array(var['Absorption-Thickness'])
y = np.array(var['LN(count)'])

plt.figure(figsize=(12,5))
plt.scatter(x, y, s=100, linewidths=1, edgecolor='black')
plt.title('ln() Plot')
plt.xlabel('Absorption Thickness', fontsize=15)
plt.ylabel('ln($N/N_0$)', fontsize=15)
plt.plot(x, y, c = 'r', linewidth=5, alpha=.5, solid_capstyle='round')
plt.scatter(x=x.mean(), y=y.mean(), marker='d', s=8**2.5, c='r') # average point
plt.savefig('ln() Plot')
```
