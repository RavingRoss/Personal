```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
var = pd.read_csv('xray-LiF.csv')

x = list(var['voltage'])
y = list(var['counts-voltage'])

plt.figure(figsize=(7,7))
plt.style.use('seaborn')
plt.scatter(x,y,s=20,edgecolors="black",c="green",label='Counts vs Voltage')
plt.xlabel('Voltage [V]',fontsize=12,fontweight='bold')
plt.ylabel('Counts [s$^{-1}$]',fontsize=12,fontweight='bold')
plt.title("Counts vs. Voltage",fontsize=15,fontweight='bold')
plt.plot(x,y,'g',linewidth=1)
plt.legend()
plt.savefig('Counts vs Voltage')
plt.show()
```

```python
var = pd.read_csv('xray.csv')

x = list(var['current'])
y = list(var['counts-current'])

plt.figure(figsize=(7,7))
plt.style.use('seaborn')
plt.scatter(x,y,s=20,edgecolors="black",c="blue",label='Counts vs Current')
plt.xlabel('Current [\u03bcA]',fontsize=12,fontweight='bold')
plt.ylabel('Counts [s$^{-1}$]',fontsize=12,fontweight='bold')
plt.title("Counts vs. Current",fontsize=15,fontweight='bold')
plt.plot(x,y,linewidth=1)
plt.legend()
plt.savefig('Counts vs Current')
plt.show()
```
