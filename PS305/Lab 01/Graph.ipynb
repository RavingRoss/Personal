# Graphs for Lab 01 Plank's Constant

```python
from ast import main
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
var = pd.read_excel('400nm LED spectra.xlsx')

y = list(var['I/I_max'])
x = list(var['Wavelength (nm)'])

plt.figure(figsize=(7,7))
plt.style.use('seaborn')
plt.scatter(x,y,s=20,edgecolors="black",c="purple",label='400nm Curve')
plt.xlabel('Wavelength [nm]',fontsize=12,fontweight='bold')
plt.ylabel('Intensity [W/m$^{2}$]',fontsize=12,fontweight='bold')
plt.title("400nm Blackbody Curve",fontsize=15,fontweight='bold')
plt.xlim(380,450)
plt.plot([403.32, 412.50], [0.5070, 0.5070], 'r', linewidth=3, label='FWHM')
plt.legend()
plt.savefig('400nm Curve')
plt.show()
```
